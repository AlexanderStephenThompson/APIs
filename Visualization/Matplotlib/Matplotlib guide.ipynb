```python
# Sample Scatterplot

import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.random.rand(50)  # Random x values
y = np.random.rand(50)  # Random y values

# Create a scatter plot
plt.scatter(x, y, label='Random Data', color='b', marker='o', s=50)

# Add labels and a legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Scatter Plot')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
