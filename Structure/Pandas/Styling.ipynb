````python
import pandas as pd
import numpy as np

'''
Resources
 - https://www.youtube.com/shorts/4MAvYHD7tz4
 - https://www.youtube.com/watch?v=JGefS6WPm1E
 - https://www.youtube.com/watch?v=bBbxumJakuA
'''
````

````python
np.random.seed(0)

df = pd.DataFrame(np.random.randn(10,4),columns=['A', 'B', 'C', 'D'])
df
````

````python
null_df = pd.DataFrame(np.random.randn(10,4),columns=['A', 'B', 'C', 'D'])
null_df.iloc[0,2] = np.nan
null_df.iloc[4,3] = np.nan

null_df.style.highlight_null(color="red").format(na_rep='<span style="color:white">NaN</span>')
````

````python
import seaborn as sns

heatmap_coloring = sns.light_palette("green", as_cmap=True)

df.style.background_gradient(cmap=heatmap_coloring)
````

````python
def style_column_D(cell):
    return f'color: white;'

diverging_colors = sns.color_palette("RdYlGn", as_cmap=True)
df.style.bar(subset=['D'], cmap=diverging_colors, align='zero').applymap(style_column_D, subset=pd.IndexSlice[:, ['D']])
````

````python
df.style.highlight_max(subset=['D'], props='background-color: green; color:white;')
````

````python
df.style.highlight_min(subset=['D'], props='background-color: red; color:white;')
````

````python
# You can highlight multiple things.

df.style\
    .highlight_max(subset=['D'], props='background-color: green; color:white;')\
    .highlight_min(subset=['D'], props='background-color: red; color:white;')
````

````python
# Highlight a range of values
df.style.highlight_between(subset=['D'], left=0, right=3, props='background-color: green; color:white;')
````

## Formatting

```python
df.style.format({
    'A': '{:.2%}',        # Percentage format
    'B': '${0:,.2f}',     # Currency format
    'C': '{:.2e}',        # Scientific notation format
    'D': '{:,.0f}',       # Integer format with thousands separator
    'E': 'Number: {:.1f}',# Custom numeric format
    'F': '{:%Y-%m-%d}'    # Date format (assuming datetime objects in column 'F')
})
```

````python
df.style.format({'A': '{:.2%}',     # Percentage
                 'B': '${0:,.2f}'}  # Currency
                )
````
